```python
import pandas as pd
from pandas import Series, DataFrame
import folium
from folium import plugins
```

```python
path = './dataset/'
endName = {'2017','2018'}
filenames = {'Janeiro', 'Fevereiro','Março','Abril','Maio','Junho','Junho', 'Julho', 'Agosto', 'Setembro'}
extension = '.xlsx'
```

```python
def populaDataFrame():
    dfList = list()

    for ano in endName:
        for file in filenames:
            df= pd.read_excel(path+file+ano+extension)
            dfList.append(df)

    df = pd.concat(dfList)
    return df
    
    
```

```python
df = populaDataFrame()
df.reset_index()
df.drop_duplicates(['NUM_BO'])
```

```python
df17 = df.loc[(df['ANO_BO'] == 2017)]
df17 = df17[['LATITUDE', 'LONGITUDE']]
df17.dropna(inplace=True)
```

```python
df18 = df.loc[(df['ANO_BO'] == 2018)]
df18 = df18[['LATITUDE', 'LONGITUDE']]
df18.dropna(inplace=True)
```

```python
coordenadas17 = []
lat = df17['LATITUDE'].values
long = df17['LONGITUDE'].values

mapa = folium.Map(location=[-23.1791, -45.8872],tiles='OpenStreetMap',zoom_start=12)
 
for la,lo in zip(lat,long):
    coordenadas17.append([la,lo])

mapa.add_child(plugins.HeatMap(coordenadas17))
```
